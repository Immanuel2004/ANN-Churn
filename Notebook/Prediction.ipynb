```python
import pickle
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import load_model
```

```python
# Load the Trained Model and All the Pickle Files 

Model = load_model('Model.h5')

# Loading the Pickle files
with open('Label_Encoder_Gender.pkl','rb') as file:
    Encoded_Gender = pickle.load(file)

with open('OHE_Geography.pkl','rb') as file:
    Encoded_Geography = pickle.load(file)

with open('Scaler.pkl','rb') as file:
    Scaler = pickle.load(file)
```

```python
input_data = {
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':600000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
Geo_Encoded = Encoded_Geography.transform([[input_data['Geography']]]).toarray()
Geo_Encoded_df = pd.DataFrame(Geo_Encoded,columns=Encoded_Geography.get_feature_names_out(['Geography']))
```

```python
Geo_Encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender'] = Encoded_Gender.transform(input_df['Gender'].values)
input_df
```

```python
# Concatination
input_df = pd.concat([input_df.drop('Geography',axis=1),Geo_Encoded_df],axis=1)
input_df
```

```python
input_scaled = Scaler.transform(input_df)
input_scaled
```

```python
Prediction = Model.predict(input_scaled)
Prediction
```
